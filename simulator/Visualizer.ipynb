```python
import numpy as np
import ast
import matplotlib.pyplot as plt
%matplotlib inline

ny, nx = 10, 10

all_data = None
with open("data/result/resultUpd", 'r') as all_data:
    print "AAA"
    all_data = ast.literal_eval(all_data.read().strip())
    print "BBB"

curr_it = 1
for in_data in all_data:
    r = np.ones(10*10).reshape(10, 10)
    g, b = [(np.ones(ny*nx) / 1.).reshape((ny, nx)) for _ in range(2)]

    for i in range(10):
        r[i][i] = 0.
        g[i][i] = 0.
        b[i][i] = 0.

    for cell_cords in in_data:
        tmp = ast.literal_eval(cell_cords)
        r[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])
        g[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])
        b[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])

    c = np.dstack([r,g,b])

    plt.imshow(c, interpolation='nearest')
    plt.show()
    plt.savefig("simulator" + str(curr_it) + ".jpg")
    curr_it += 1
```
