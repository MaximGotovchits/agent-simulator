```python
import numpy as np
import ast
import matplotlib.pyplot as plt
%matplotlib inline

nx,ny = 46, 61

all_data = None
with open("data/result/resultRealData", 'r') as all_data:
    all_data = ast.literal_eval(all_data.read().strip())

curr_it = 1
for in_data in all_data:
    r = np.ones(nx*ny).reshape(nx, ny)
    g, b = [(np.ones(ny*nx) / 1.).reshape((nx, ny)) for _ in range(2)]

    for i in range(nx):
        for j in range(ny):
            r[i][j] = 1.
            g[i][j] = 1.
            b[i][j] = 1.

    for cell_cords in in_data:
        tmp = ast.literal_eval(cell_cords)
        r[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])
        g[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])
        b[int(tmp[0])][int(tmp[1])] = float(in_data[cell_cords])

    c = np.dstack([r,g,b])

    plt.imshow(c, interpolation='nearest')
    plt.savefig('data/visual_result_40meteres/simulator' + str(curr_it) + '.png')
    plt.show()
    curr_it += 1
```
